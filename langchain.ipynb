```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage

llm = ChatOpenAI(model="gpt-4o-mini", verbose=True)

history = []

while True:
    user_input = input("User: ")
    
    if user_input == "exit":
        print("Exiting the chat.")
        break
    
    history.append(HumanMessage(content=user_input))
    
    response = llm.invoke(history)
    
    history.append(AIMessage(content=response.content))
    
    print("User Input:", user_input)
    
    print("LLM Response:", response.content)
    
    
    
```

```python
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document

# Sample documents (your knowledge base)
docs = [
    Document(page_content="Bennu is a snowflake engineer. Who is good in painting isro"),
    Document(page_content="Ragul is a software engineer. He is good in cricket"),
    Document(page_content="Guna is a data scientist. He is working in isro"),
    Document(page_content="Siva is a software engineer. He is good in football"),   
    Document(page_content="Shenba is a Player. He is good in volleyball")
]

# Create embeddings
embeddings = OpenAIEmbeddings()

# Create vector store
vectorstore = FAISS.from_documents(docs, embeddings)
```

```python
user_input = "isro"

retrieved_docs = vectorstore.similarity_search(user_input, k=3)
context = "\n".join([doc.page_content for doc in retrieved_docs])

print(context)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage

llm = ChatOpenAI(model="gpt-4o-mini")

history = []

while True:
    user_input = input("User: ")
    
    if user_input == "exit":
        print("Exiting the chat.")
        break

    # 🔹 Retrieve relevant docs
    retrieved_docs = vectorstore.similarity_search(user_input, k=3)
    context = "\n".join([doc.page_content for doc in retrieved_docs])
    
    
    print("Retrieved Context: " + context)

    # 🔹 Build prompt with context
    prompt = f"""
    Use the following context to answer:

    {context}

    Question: {user_input}
    """
    
    
    print("Prompt to LLM: " + prompt)

    history.append(HumanMessage(content=prompt))

    response = llm.invoke(history)

    history.append(AIMessage(content=response.content))

    print("AI Response:", response.content)

print(history)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage

llm = ChatOpenAI(model="gpt-4o-mini")

history = []

while True:
    user_input = input("User: ")
    
    if user_input == "exit":
        print("Exiting the chat.")
        break

    prompt = f"""
    Use the following context to answer:

    My name shiva and i am a software engineer. I have experience in python and java. Overall i have 5 years of experience in software development. I have worked on various projects including web development, data analysis and machine learning. I am passionate about learning new technologies and improving my skills.

    
    Question: {user_input}
    """
    
    print(prompt)

    history.append(HumanMessage(content=prompt))

    response = llm.invoke(history)

    history.append(AIMessage(content=response.content))

    print("AI Response:", response.content)

print(history)
```
